## Q1 — Which team won the most matches?

```python
home_wins = df[df["home_score"] > df["away_score"]]["home_team"]

away_wins = df[df["away_score"] > df["home_score"]]["away_team"]

wins = pd.concat([home_wins,away_wins]).value_counts()
print(wins)
```

## Q2 — Which World Cup had the most goals?

```python
df["total_goals"] = df["home_score"] + df["away_score"]

most_goals_worlcup = df.groupby("Year")["total_goals"].sum()

most_goals_worlcup = most_goals_worlcup.sort_values(ascending=False).head(10)
print(most_goals_worlcup)
```

## Q3 — Biggest victory margin

```python
df["goal_difference"] = (df["home_score"] - df["away_score"]).abs()

max_goal_difference = df.loc[df["goal_difference"].idxmax()]
print(max_goal_difference)
```

## Q4 — Which team received the most red cards?


# Note: counts matches with at least one red card, not total red cards

```python
home_reds = df[df["home_red_card"] != 0][["home_team", "home_red_card"]].rename(columns={"home_team": "team"})
away_reds = df[df["away_red_card"] != 0][["away_team", "away_red_card"]].rename(columns={"away_team": "team"})

red_cards = pd.concat([home_reds, away_reds])
red_cards["team"].value_counts

most_red_cards = red_cards["team"].value_counts()
most_red_cards.head(10)
```

## Q5 — How many matches ended 0–0?


```python
df[df["total_goals"] == 0].shape[0]
```

## Q6 — Which World Cup had the highest attendance?

```python
most_attendance_worlcup = df.groupby("Year")["Attendance"].sum()

most_attendance_worlcup = most_attendance_worlcup.sort_values(ascending=False).head(5)
print(most_attendance_worlcup)
```

## Q7 — Which team lost the most matches?


```python
home_losses = df[df["home_score"] < df["away_score"]]["home_team"]

away_losses = df[df["away_score"] < df["home_score"]]["away_team"]

losses = pd.concat([home_losses, away_losses]).value_counts()
print(losses)
```

```python
df["Round"].unique()
```

```python
knockout = df[~df["Round"].isin(['Group stage', 'First group stage', 'Second group stage', 'First round', 'Second round', 'Group stage play-off'])]
knockout.shape[0]
```

## Q8 — What percentage of knockout matches were decided by penalties?

```python
shootouts = knockout["home_penalty_shootout_goal_long"].notna().sum()
total_knockout = knockout.shape[0]
percentage = (shootouts / total_knockout) * 100

print(f"Partidos con tanda: {shootouts}")
print(f"Total eliminatorias: {total_knockout}")
print(f"Porcentaje: {percentage:.1f}%")
```

## Q9 — Correlation between yellow cards and goals

```python
df["home_yellows"] = df["home_yellow_card_long"].str.count("&rsquor;").fillna(0)
df["away_yellows"] = df["away_yellow_card_long"].str.count("&rsquor;").fillna(0)
df["total_yellows"] = df["home_yellows"] + df["away_yellows"]

df[["total_yellows", "total_goals"]].corr()
```

## Final Report — Export to CSVDijiste: clauede, esto si esta bein no son datos muy alejados?

```python
summary = pd.DataFrame({
    "metric": [
        "total_matches",
        "total_goals",
        "avg_goals_per_match",
        "biggest_victory_margin",
        "top_winning_team",
        "most_red_cards_team",
        "zero_goal_matches",
        "knockout_penalty_pct"
    ],
    "value": [
        df.shape[0],
        df["total_goals"].sum(),
        round(df["total_goals"].mean(), 2),
        df["goal_difference"].max(),
        wins.index[0],
        most_red_cards.index[0],
        df[df["total_goals"] == 0].shape[0],
        round(percentage, 1)
    ]
})

summary
```

```python
df["decade"] = (df["Year"] // 10) * 10

by_decade = df.groupby("decade").agg(
    total_matches=("total_goals", "count"),
    total_goals=("total_goals", "sum")
).reset_index()

by_decade
```

```python
summary.to_csv("output/wc_summary.csv", index=False)
by_decade.to_csv("output/wc_by_decade.csv", index=False)

print("Exportado")
```
